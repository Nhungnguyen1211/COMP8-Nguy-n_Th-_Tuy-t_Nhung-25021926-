```python
#W8A1
class Cylinder:
  def __init__(self, r, h):
    self.r = r
    self.h = h
  def stp(self):
    return (2 * 3.14 * self.r * self.h) + 2 *( 3.14 * self.r **2)
  def thetich(self):
    return ( 3.14 * self.r **2 * self.h)
r,h = map(int, input().split())
cyl= Cylinder (r,h)
print(f'{cyl.stp():.2f}')
print(f'{cyl.thetich():.2f}')



```

```python
#W8A2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0)

    def days_in_month(self):
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        if self.month in [4, 6, 9, 11]:
            return 30
        if self.month == 2:
            return 29 if self.is_leap_year() else 28
        return 0

    def is_valid(self):
        if self.year <= 0:
            return False
        if self.month < 1 or self.month > 12:
            return False
        if self.day < 1 or self.day > self.days_in_month():
            return False
        return True

    def next_date(self):
        if not self.is_valid():
            return None

        d, m, y = self.day, self.month, self.year

        if d < self.days_in_month():
                        d += 1
        else:
            d = 1
            if m < 12:
                m += 1
            else:
                m = 1
                y += 1

        return Date(d, m, y)

    def __str__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year:04d}"
s = input().strip()
day, month, year = map(int, s.split("/"))

date = Date(day, month, year)

if not date.is_valid():
    print("INVALID")
else:
    print(date.next_date())

```

```python
#W8A3
import math
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def kt (self):
    if self.x == 0 and self.y != 0:
      return " điểm thuộc trục tung"
    elif self.x != 0 and self.y == 0:
      return "điểm thuộc trục hoành"
    elif self.x == 0 and self.y == 0:
      return "điểm là gốc tọa độ"
    else:
      return "điểm không thuộc các trục"
  def distance(self):
    return (math.sqrt(self.x**2 + self.y**2))
x,y=map(int,input().split())
M=Point(x,y)
print(M.kt())
print(f'khoảng của điểm đến gốc tọa độ là: {M.distance():.2f}')

```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.products = []
    def add_product(self, name, price):
        self.products.append({"name": name, "price": price})
    def remove_product(self, name):
        for product in self.products:
            if product["name"] == name:
                self.products.remove(product)
                return
        print("Product not found")
    def is_empty(self):
        return len(self.products) == 0
    def total_price(self):
        total = 0
        for product in self.products:
            total += product["price"]
        return total
    def show_products(self):
        if self.is_empty():
            print("Cart is empty")
            return
        for product in self.products:
            print(f"{product['name']} - {product['price']}")
    def clear_cart(self):
        self.products.clear()
cart = ShoppingCart()
while True:
    print("\n--- MENU ---")
    print("2. Add product")
    print("3. Remove product")
    print("4. Check empty")
    print("5. Total price")
    print("6. Show products")
    print("7. Clear cart")
    choice = input("Choose (2-7): ")
    if choice == "2":
        name = input("Product name: ")
        price = float(input("Price: "))
        cart.add_product(name, price)
    elif choice == "3":
        name = input("Product name to remove: ")
        cart.remove_product(name)
    elif choice == "4":
        if cart.is_empty():
            print("Cart is empty")
        else:
            print("Cart is not empty")
    elif choice == "5":
        print("Total price:", cart.total_price())
    elif choice == "6":
        cart.show_products()
    elif choice == "7":
        cart.clear_cart()
        print("Cart cleared")
    else:
        print("Invalid choice")
    exit_choice = input("Có muốn thoát khỏi chương trình không? ")
    if exit_choice.lower() == "yes":
        break




```

```python
#W8A6
class Rectangle:
  def __init__(self,w,h):
    self.w=w
    self.h=h
  def area(self):
    return self.w*self.h
  def perimeter(self):
    return 2*(self.w+self.h)
  def scale(self,k):
    self.w*=k
    self.h*=k
w,h,k=map(int,input().split())
rect=Rectangle(w,h)
rect.scale(k)
print(f"{rect.area():.2f} {rect.perimeter():.2f}")

```

```python
#W8A8
class BankAccount:
  def __init__(self,owner,balance):
    self.owner=owner
    self.balance=balance
  def deposite(self,amount):
    self.balance+=amount
    return self.balance
  def withdraw(self,amount):
    if amount>self.balance:
      return ("Số dư không đủ")
    else:
      self.balance-=amount
      return self.balance
  def __str__(self):
    return(f"{self.owner}:{self.balance}")
line=input().strip()
owner,balance=line.split(" ",1)
account=BankAccount(owner,float(balance))
n=int(input())
for i in range(n):
  cmd,x=input().split()
  x=float(x)
  if cmd=="DEPOSIT":
    account.deposite(x)
  if cmd=="WITHDRAW":
    account.withdraw(x)
print(f"{account.balance:.2f}")

```

```python
#W8A9
class Student:
  def __init__(self,name):
    self.name=name
    self.scores=[]
  def add_score(self,subject,score):
    self.scores.append(score)
  def avg(self):
    return sum(self.scores)/len(self.scores)
  def rank(self):
    if self.avg()>=8:
      return("Excellent")
    elif self.avg() >=6.5:
      return("Good")
    elif self.avg()>=5:
      return("Average")
    else:
      return("Poor")
name=input().strip()
student=Student(name)
n=int(input())
for i in range (n):
  line=input().strip()
  subject,score=line.split(" ",1)
  student.add_score(subject,float(score))
print(f"{student.name} {student.avg():.2f} {student.rank()}")
```

```python
#W8A10
class Book:
  def __init__(self,title,author,year):
    self.title=title
    self.author=author
    self.year=year
class Library:
  def __init__(self):
     self.books=[]
  def add(self,book):
    self.books.append(book)
  def count_by_author(self,name):
    count=0
    for book in self.books:
      if book.author == name:
        count+=1
    return count
  def find_by_year(self,y):
    count=0
    for book in self.books:
      if book.year==y:
        count+=1
    return count
library = Library()
n=int(input())
for i in range(n):
  line=input().strip()
  if line.startswith("ADD"):
    _,data=line.split(" ",1)
    title,author,year=data.split(";")
    book=Book(title,author,int(year))
    library.add(book)
  elif line.startswith("COUNTYEAR"):
    _,year=line.split()
    print(library.find_by_year(int(year)))
  elif line.startswith("COUNT"):
    _,author=line.split(" ",1)
    print(library.count_by_author(author))




```

```python
#W8A11
import math
class Point:
  def __init__(self,x , y, z):
    self.x = x
    self.y = y
    self.z = z
  def vecto_to(self, other):
    return Vector(other.x-self.x,
                  other.y-self.y,
                  other.z-self.z)
class Vector(Point):
  def dot(self,other):
    return (self.x * other.x + self.y * other.y + self.z * other.z)
  def cross(self,other):
    return Vector(self.y*other.z-self.z*other.y,self.z*other.x-self.x*other.z,self.x*other.y-self.y*other.x)
  def magnitude(self):
    return math.sqrt(self.x**2+self.y**2+self.z**2)
points=[]
for i in range(4):
  x,y,z=map(int,input().split())
  points.append(Point(x,y,z))
A,B,C,D=points
n1=(B.vecto_to(A)).cross((B.vecto_to(C)))
n2=(B.vecto_to(A)).cross((B.vecto_to(D)))
cos_phi=(n1.dot(n2))/(n1.magnitude()*n2.magnitude())
a=math.degrees(math.acos(cos_phi))
if a>90:
  b=180-a
  print(f"{b:.2f}")
else:
  print(f"{a :.2f}")








```

```python
#W8A13
import math
class Triangle:
  def __init__(self,egde1,egde2,egde3):
    self.egde1=egde1
    self.egde2=egde2
    self.egde3=egde3
  def check(self):
    if (self.egde1+self.egde2<=self.egde3) or (self.egde2+self.egde3<= self.egde1) or (self.egde1+self.egde3<=self.egde2):
      return "invalid"
    self.p=(self.egde1+self.egde2+self.egde3)/2
    return math.sqrt(self.p*(self.p-self.egde1)*(self.p-self.egde2)*(self.p-self.egde3))
egde1,egde2,egde3=map(int,input().split())
tri=Triangle(egde1,egde2,egde3)
print(tri.check())
```
