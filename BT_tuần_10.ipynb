```python
#W8A11
def is_valid_regex(s):
    repeats = {'*', '+', '?'}
    prev = ''
    for c in s:
        if c in repeats and prev in repeats:
            return False
        if c == ')' and prev == '(':
            return False
        prev = c
    return True
t = int(input())
for i in range(t):
    s = input()
    print("Valid" if is_valid_regex(s) else "Invalid")
```
