```python
#W4A7
n = int(input())
largest = 0
# Loại bỏ ước số 2
while n % 2 == 0:
    largest = 2
    n //= 2
# Kiểm tra các ước số lẻ
i = 3
while i * i <= n:
    while n % i == 0:
        largest = i
        n //= i
    i += 2
# Nếu n còn lại > 1 thì nó là số nguyên tố lớn nhất
if n > 1:
    largest = n
print(largest)

```

```python
#W4A8
def reverse_num(x):
    return int(str(x)[::-1])
def is_palindrome(x):
    s = str(x)
    return s == s[::-1]
n = int(input())
steps = 0
while not is_palindrome(n):
    n = n + reverse_num(n)
    steps += 1
print(steps, n)
```

```python
#W4A9
n = int(input())
result = []
def distinct_digits(x):
    s = str(x)
    return len(s) == len(set(s))
i = 1
while i * i <= n:
    sq = i * i
    if distinct_digits(sq):
        result.append(sq)
    i += 1
if result:
    print(*result)
else:
    print("no number")
```

```python
#W4A10
n = int(input())

collatz_len = {1: 1}

def get_length(x):
    if x in collatz_len:
        return collatz_len[x]
    if x % 2 == 0:
        nxt = x // 2
    else:
        nxt = 3 * x + 1
    collatz_len[x] = 1 + get_length(nxt)
    return collatz_len[x]

best_x = 1
best_len = 1

for x in range(1, n + 1):
    L = get_length(x)
    if L > best_len or (L == best_len and x < best_x):
        best_len = L
        best_x = x

print(best_x, best_len)
```

```python
#W4A11
n = int(input())

count = 0
i = 1

while i * i <= n:
    if n % i == 0:
        # i là ước
        if i % 2 == 0:
            count += 1
        # n // i là ước còn lại (khác i)
        if i * i != n and (n // i) % 2 == 0:
            count += 1
    i += 1

print(count)
```

```python
#W4A13
def sum_divisors(x):
    total = 1  # 1 luôn là ước số
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            total += i
            if i != x // i:
                total += x // i
    return total

a = int(input())
b = int(input())

if sum_divisors(a) == b + a and sum_divisors(b) == a + b:
    print("true")
else:
    print("false")
```

```python
#W4A32
password = input("Nhập mật khẩu: ")

has_upper = False
has_lower = False
has_digit = False
has_special = False

special_chars = "!@#$%^&*()-_=+[]{},.<>/?;:'\"\\|"

for ch in password:
    if ch.isupper():
        has_upper = True
    elif ch.islower():
        has_lower = True
    elif ch.isdigit():
        has_digit = True
    elif ch in special_chars:
        has_special = True

if len(password) > 6 and has_upper and has_lower and has_digit and has_special:
    print("→ Mật khẩu mạnh")
else:
    print("→ Mật khẩu yếu")
```
