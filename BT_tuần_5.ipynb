```python
#W5A3
import math
def isPrime(n):
    for i in range(2,int(math.sqrt(n))):
        if n%i==0:
            return False
    return True
print(isPrime(2))
```

```python
#W5A4
def perfect_number(n):
  m=0
  for i in range (1,n):
    if n%i==0:
      m=m+i
  if m==n:
    return True
  return False
print(perfect_number(28))
```

```python
#W5A5
def index(nums,k):
  for i in range(len(nums)):
    if nums[i]==k:
      return i+1
  return -1
print(index([1,3,45,566,675,65,6,54],6))
```

```python
#W5A6
def fac(n):
  if n<0:
    print("nhập lại")
  elif n==0 or n==1:
    return 1
  else:
    result=1
    for i in range(2,n+1):
       result=result*i
    return result
print(fac(4))
```

```python
#W5A7
def calculator(num1, num2, operat):
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        if num2 != 0:
            result = num1 / num2
        else:
            return "Error: Division by zero"
    else:
        return "Error: Invalid operator"

    return round(result, 2)
print(calculator(2,3,'+'))
```

```python
#W5A8
def haming(x:int, y:int) -> int:
    xor = x^y
    distance=0
    while xor:
        distance += xor&1
        xor >>=1
    return distance
print(haming(1,4))
```

```python
#W5A10
def are_isomorphic(a, b):
    if len(a) != len(b):
        return False

    mapping_a_to_b = {}
    mapping_b_to_a = {}

    for char_a, char_b in zip(a, b):
        if char_a in mapping_a_to_b:
            if mapping_a_to_b[char_a] != char_b:
                return False
        else:
            mapping_a_to_b[char_a] = char_b

        if char_b in mapping_b_to_a:
            if mapping_b_to_a[char_b] != char_a:
                return False
        else:
            mapping_b_to_a[char_b] = char_a

    return True
print(are_isomorphic("egg","add"))
```
